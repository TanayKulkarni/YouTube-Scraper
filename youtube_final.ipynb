```python
x="https://www.youtube.com/playlist?list=PLEbnTDJUr_IegfoqO4iPnPYQui46QqT0j"


url = (x)
r = requests.get(url)
#r.content


soup = BeautifulSoup(r.content)
# print (soup.prettify())


channel_name1=soup.find_all("a",{"class": " yt-uix-sessionlink spf-link "})
channel4 = channel_name1[1].text


videos = soup.find_all("a",{"class": "pl-video-title-link yt-uix-tile-link yt-uix-sessionlink spf-link "})
vids = soup.find_all('a',attrs={'class':'yt-uix-tile-link'})
# lis=[]
i=0
while(i<len(videos)):
        
        lis.append(videos[i].text)
#         print(videos[i].text)

        i+=1
# lis = map(lambda each:each.strip("\n"), lis)
# lis = ([s.strip('\n') for s in lis])
# lis = ([s.replace('\n', '') for s in lis])
# lis = ([s.replace(' ', '') for s in lis])
# lis

page = r.text
soup=BeautifulSoup(page,'html.parser')
vids = soup.findAll('a',attrs={'class':'yt-uix-tile-link'})

# videolist=[]
for v in vids:
    tmp = 'https://www.youtube.com' + v['href']
    videolist.append(tmp)

lis = ([s.replace('\n', '') for s in lis])
lis = ([s.replace(' ', '') for s in lis])
# len(videolist)
```
